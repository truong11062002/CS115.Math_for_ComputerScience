# Chia tập dữ liệu bằng  train_test_split

```python
X = penguins.iloc[:,[2,4]].values
y = penguins.iloc[:,0].values
```

```python
# Chuẩn hóa kì vọng = 0, phương sai = 1
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X = SC.fit_transform(X)
```

```python
# Chia dữ liệu thành 2 tập training set và test set
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X,y,train_size = 0.8, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_4maxdepth = DecisionTreeClassifier(max_depth=4, random_state=42)
model_4maxdepth.fit(X_train,y_train)
y_pred = model_4maxdepth.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_2maxdepth = DecisionTreeClassifier(max_depth=2, random_state=42)
model_2maxdepth.fit(X_train,y_train)
y_pred_2 = model_2maxdepth.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred, y_test))
print(accuracy_score(y_pred_2, y_test))
```

```python
from graphviz import Source
from sklearn.tree import export_graphviz
```

# Xây dựng model Decision Tree

```python
export_graphviz(model_2maxdepth, out_file='penguins.dot',feature_names=['bill_length_mm','flipper_length_mm'],
                class_names=['Adelie','Chinstrap', 'Gentoo'],impurity=False,
                filled=True, rounded=True)
Source.from_file('penguins.dot')
```

```python
from sklearn.preprocessing import StandardScaler
bill_length_sc = StandardScaler()
flipper_length_sc = StandardScaler()
bill_length = bill_length_sc.fit_transform(penguins.iloc[:, 2].values.reshape(-1, 1))
flipper_length = flipper_length_sc.fit_transform(penguins.iloc[:, 4].values.reshape(-1, 1))
penguins.iloc[:, 2] = bill_length.flatten()
penguins.iloc[:, 4] = flipper_length.flatten()
sns.jointplot(data = penguins, x = penguins.columns[2], y = penguins.columns[4], hue = 'species', palette = ['orange', 'green', 'purple'])
```

```python
export_graphviz(model_4maxdepth, out_file='penguins_4maxdepth.dot',feature_names=['bill_length_mm','flipper_length_mm'],
                class_names=['Adelie','Chinstrap', 'Gentoo'],impurity=False,
                filled=True, rounded=True)
Source.from_file('penguins_4maxdepth.dot')
```

```python
score = []
depth = []
def model():
  for i in range(1, 11):
    model = DecisionTreeClassifier(max_depth=i, random_state=42)
    model = model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score.append(accuracy_score(y_pred, y_test))
    depth.append(i)
model()
```

```python
visual = pd.DataFrame(data = score, columns = ['Accuracy_score'])
visual.insert(0, 'Max_depth', depth)
visual
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(visual.iloc[:, 0], visual.iloc[:, 1], color = 'red')
plt.plot(visual.iloc[:, 0], visual.iloc[:, 1], color = 'blue')
plt.title('Accuracy Score vs Max Depth')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy Score (%)')
plt.show()
```

# Nhận xét

Sau khi khảo sát sự tương quan giữa max_depth và độ đo Accuracy, nhận thấy được rằng lúc đầu, độ sâu càng cao thì độ đo cũng càng cao nhưng kể từ khi tăng max_depth từ 4 trở lên thì model ngày càng phức tạp hơn, không chỉ dẫn tới tình trạng overfitting mà còn thậm chí làm giảm độ chính xác của model

Từ đó, việc chọn một max_depth khôn ngoan sẽ quyết định lớn đến cả model đang xây dựng, và tất nhiên không phải cứ càng phức tạp là càng tốt, mà nó còn tệ hơn lúc đầu.

Bên cạnh đó, việc độ đo Accuracy không thay đổi từ max_depth = 7 trở đi cho thấy là cây chỉ có độ sâu tối đa là 7, nói cách khác, đó là khi model phức tạp nhất có thể trong mô hình Cây quyết định với bộ dữ liệu này
