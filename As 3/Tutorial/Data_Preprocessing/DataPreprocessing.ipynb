```python
Q1 = df_1['Height'].quantile(0.25)
Q3 = df_1['Height'].quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
IQR = Q3 - Q1
low_fence = Q1 - 1.5*IQR
up_fence = Q3 + 1.5*IQR
```

```python
df_1[((df_1['Height'] < low_fence) | (df_1['Height'] > up_fence))]
```

```python
df_1 = df_1[~((df_1['Height'] < low_fence) | (df_1['Height'] > up_fence))]
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder(sparse=False)
```

```python
name_col = 'Grade'
```

```python
data_new = onehot_encoder.fit_transform(df_1[[name_col]])
```

```python
grade_column = pd.DataFrame(data=data_new, columns=onehot_encoder.get_feature_names([name_col]))
```

```python
df_1 = pd.concat([df_1.drop(columns=[name_col, 'Good-looking']), grade_column, df_1['Good-looking']], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_1.drop(columns=['ID', 'Good-looking'])
y = df_1['Good-looking']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
```

```python
X_train[['Height']] = min_max_scaler.fit_transform(X_train[['Height']])
X_test[['Height']] = min_max_scaler.fit_transform(X_test[['Height']])
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
```

```python
X_train[['Age']] = standard_scaler.fit_transform(X_train[['Age']])
X_test[['Age']] = standard_scaler.fit_transform(X_test[['Age']])
```
